### DATA (BEFORE) - NULL VALUE IMPUTATION

```python
data_before = pd.read_csv('Dataset-Before-Encoded.csv')
data_before.head()
```

```python
data_copy = data_before.copy()
```

```python
data_before.drop('ID',axis=1,inplace=True)
```

```python
data_before.isnull().sum()
```

```python
# percentage of null values
(data_before.isnull().sum().sum()/(data_before.shape[0]*data_before.shape[1]))*100
```

### Hot Deck Imputation - Similarity is measured by considering Nearest Neighbours

```python
# fancy impute is going to remove the column names , so storing the column names first
data_before_cols = data_before.columns
data_before_cols
```

```python
# use 5 nearest rows which have a feature to fill in each row's missing features
data_before = pd.DataFrame(KNN(k=5).fit_transform(data_before))
# this will help us to replace the values
```

```python
# here we are adding the columns back
data_before.columns = data_before_cols
```

```python
# check for null values
data_before.isnull().sum()
```

```python
data_before.head()
```

```python
### rounding off values
def roundvals(dataframe):
    for i in dataframe.columns:
        dataframe[i] = dataframe[i].apply(lambda x:round(x))
```

```python
roundvals(data_before)
```

```python
data_before.head()
```

```python
data_before_cols
```

```python
data_copy.drop(data_before_cols,axis=1,inplace=True)
```

```python
data_before.head()
```

```python
data_copy.shape,data_before.shape
```

```python
data_before_final = pd.concat([data_copy,data_before],axis=1)
```

```python
data_before_final.head()
```

```python
data_before.isnull().sum()
```

```python
data_before_final.to_csv('Dataset-Before-HotDeck.csv')
```
